# Customer Segmentation with Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

np.random.seed(42)
```

## What is Customer Segmentation?

Customer segmentation is grouping customers into meaningful categories based on behavior or attributes. In an unsupervised setting, you do not have labels like “premium” or “churn risk.” Instead, you let patterns in the data suggest segments.

A clustering model does not automatically create business meaning. The clustering step produces groups. Interpretation comes from what the features represent, how stable the clusters are, and whether the segments map to actions you can take.

```python
# fake dataset

n = 600
segment = np.random.choice([0, 1, 2], size=n, p=[0.45, 0.35, 0.20])


recency = np.where(segment==0, np.random.gamma(2, 10, n),
np.where(segment==1, np.random.gamma(2, 25, n),
np.random.gamma(2, 7, n)))


frequency = np.where(segment==0, np.random.poisson(4, n),
np.where(segment==1, np.random.poisson(2, n),
np.random.poisson(10, n)))


monetary = np.where(segment==0, np.random.lognormal(3.3, 0.3, n),
np.where(segment==1, np.random.lognormal(3.0, 0.35, n),
np.random.lognormal(3.8, 0.25, n)))


customers = pd.DataFrame({
"customer_id": np.arange(n),
"recency_days": recency,
"frequency": frequency,
"monetary": monetary
})


customers.head()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 4))
ax[0].hist(customers["recency_days"], bins=30)
ax[0].set_title("Recency (days)")
ax[1].hist(customers["frequency"], bins=30)
ax[1].set_title("Frequency")
ax[2].hist(customers["monetary"], bins=30)
ax[2].set_title("Monetary")
plt.show()
```

```python
# scale the data

X = customers[["recency_days", "frequency", "monetary"]].copy()


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# k means Clustering

ks = range(2, 9)
inertias = []
sils = []


for k in ks:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    inertias.append(km.inertia_)
    sils.append(silhouette_score(X_scaled, labels))


plt.figure(figsize=(6, 4))
plt.plot(list(ks), inertias)
plt.title("Elbow Plot (Inertia)")
plt.xlabel("k")
plt.ylabel("Inertia")
plt.show()


plt.figure(figsize=(6, 4))
plt.plot(list(ks), sils)
plt.title("Silhouette Score vs k")
plt.xlabel("k")
plt.ylabel("Silhouette")
plt.show()
```

```python
k = 3
km = KMeans(n_clusters=k, random_state=42, n_init=10)
labels_km = km.fit_predict(X_scaled)


customers["cluster_kmeans"] = labels_km
customers[["customer_id", "cluster_kmeans"]].head()
```

```python
profile = customers.groupby("cluster_kmeans").agg({
"customer_id": "count",
"recency_days": "mean",
"frequency": "mean",
"monetary": "mean"
}).rename(columns={"customer_id": "n_customers"})


profile
```
